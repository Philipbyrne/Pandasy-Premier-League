# Cleaning Team Data

Data on each team will be prepared for future analysis.

First of all the raw .csv data is read in.

```python
import os
import pandas as pd
import pickle
```

```python
os.chdir('../data')
```

```python
os.getcwd()
```

```python
os.listdir()
```

```python
data20_21 = 'raw/20_21/teams.csv'
data21_22 = 'raw/21_22/teams.csv'
```

```python
teams20_21_df = pd.read_csv(data20_21)
teams21_22_df = pd.read_csv(data21_22)
```

Unneccessary and empty columns are removed.

```python
teams20_21_df = teams20_21_df.drop(["draw", "form", "loss","played","points","position","team_division","unavailable","win"],axis=1).set_index('id', drop=False)
teams21_22_df = teams21_22_df.drop(["draw", "form", "loss","played","points","position","team_division","unavailable","win"], axis=1).set_index('id', drop=False)
```

```python
file = open("./clean/20_21/teams20_21.pkl", "wb")
pickle.dump(teams20_21_df, file)
file.close()
```

```python
file = open("./clean/21_22/teams21_22.pkl", "wb")
pickle.dump(teams21_22_df, file)
file.close()
```
