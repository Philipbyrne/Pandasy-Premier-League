# Cleaning Fixture Data

Data on each fixture will be prepared for future analysis.

First of all the raw .csv data is read in.

```python
import os
import pandas as pd
import pickle
import numpy as np
```

```python
os.chdir('../data')
```

```python
os.getcwd()
```

```python
os.listdir()
```

```python
data20_21 = 'raw/20_21/fixtures.csv'
data21_22 = 'raw/21_22/fixtures.csv'
```

```python
fixtures20_21_df = pd.read_csv(data20_21, parse_dates = True)
fixtures21_22_df = pd.read_csv(data21_22, parse_dates = True)
```

```python
fixtures20_21_df
```

```python
fixtures20_21_df = fixtures20_21_df.drop(["finished_provisional", "provisional_start_time", "started","pulse_id","team_h_difficulty","team_a_difficulty","code"], axis=1).set_index('id', drop=False)
fixtures21_22_df = fixtures21_22_df.drop(["finished_provisional", "provisional_start_time", "started","pulse_id","team_h_difficulty","team_a_difficulty","code"], axis=1).set_index('id', drop=False)
```

```python
fixtures20_21_df
```

```python
file = open("./clean/20_21/fixtures20_21.pkl", "wb")
pickle.dump(fixtures20_21_df, file)
file.close()
```

```python
file = open("./clean/21_22/fixtures21_22.pkl", "wb")
pickle.dump(fixtures21_22_df, file)
file.close()
```
